```python
df['Promotion'] = df['Promotion'].fillna("None")
```

```python
location = df.groupby('Location').agg({
    'Quantity_Sold': 'sum',
    'Total_Revenue': 'sum',
    'Profit': 'sum'
}).reset_index()
```

```python
category = df.groupby('Category').agg({
    'Quantity_Sold': 'mean',
    'Total_Revenue': 'mean',
    'Profit': 'mean'
}).reset_index()
```

```python
df['Month'] = pd.to_datetime(df['Date']).dt.month
df.head()
```

```python
monthly_trends = df.groupby('Month').agg({
    'Quantity_Sold': 'sum',
    'Total_Revenue': 'sum',
    'Profit': 'sum'
}).reset_index()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
for col in ['Category', 'Product', 'Location', 'Weather_Condition', 'Promotion', 'Customer_Segment']:
    df[col] = df[col].astype('category')
```

```python
print (df['Category'].unique())
print (df['Location'].unique())
print (df['Promotion'].unique())
```

```python
def assign_season(month):
    if month in [12,1,2]:
        return 'Winter'
    elif month in [3,4,5]:
        return 'Spring'
    elif month in [6,7,8]:
        return 'Summer'
    else:
        return 'Fall'
df['Season'] = df['Month'].apply(assign_season)
```
